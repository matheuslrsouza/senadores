```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# ver estilos disponíveis
plt.style.available
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
df = pd.read_csv('2017_dados_limpos.csv', encoding='latin1', parse_dates=['DATA'])
```

```python
senadorDario = df[df['SENADOR'] == 'DÁRIO BERGER']
```

```python
# ver com o que o senador gasta mais

sns.countplot(data=df, x='TIPO_DESPESA')

plt.title('Quantidade de despesas')
plt.xlabel('Tipo de despesa')
plt.ylabel('Quantidade')
plt.show()
```

```python
valores_por_tipodespesa = senadorDario.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum()
valores_por_tipodespesa
```

```python
valores_por_tipodespesa.plot(kind='line', color='red', title='Valores gastos por tipo de despesa')
plt.show()
```

```python
# parecido com o group by, conta quantas despesas cada senador teve
df['SENADOR'].value_counts().plot(kind='bar')
plt.show()
```

```python
senadores_que_mais_gastaram = df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
senadores_que_mais_gastaram.plot(kind='bar', color='green')
plt.show()

senadores_que_mais_gastaram
```

```python
df['DATA'].dropna(inplace=True)
```

```python
# converte data para ano e mes para poder agrupar
df['DATA'] = df['DATA'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
gastos_por_mes = df.pivot_table(index='DATA', values='VALOR_REEMBOLSADO', aggfunc=np.sum)
gastos_por_mes.plot(kind='bar')
plt.plot()
```

```python
media = df.pivot_table(index='DATA', values='VALOR_REEMBOLSADO', aggfunc=np.mean)
media.columns = ['MEDIA']
mediana = df.pivot_table(index='DATA', values='VALOR_REEMBOLSADO', aggfunc=np.median)
mediana.columns = ['MEDIANA']

media_mediana = pd.concat([media, mediana], axis=1)

media_mediana
```

```python
media_mediana.plot()
plt.show()
```

```python
dados2017 = df[(df['DATA'] >= '2017-01') & (df['DATA'] <= '2017-12')]
dados2017.tail()
```

```python
mapa_de_calor = dados2017.pivot_table(index='DATA', values='VALOR_REEMBOLSADO', columns='TIPO_DESPESA', aggfunc=np.sum)
mapa_de_calor
```

```python
sns.heatmap(mapa_de_calor, annot=True, fmt='g')
plt.show()
```
